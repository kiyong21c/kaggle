# 주성분 분석(PCA, principal Component Analysis)
 - 비지도 학습
 - 예측 X, 분류 X
 - 변수의 개수를 줄이되 가능한 그 특성을 보존하는 기법
 - 기존 변수들의 정보를 모두 반영하는 새로운 변수들을 만드는 방식의 차원 축소
 - 장점 : 시각화에 유용, 변수간의 높은 상관관계 문제 해결
 - 단점 : 새로운 변수를 사용하므로 해석에 어려움, 차원축소에 따른 불가피한 정보손실

# 데이터의 차원을 축소하여 이해하기 쉽게 시각화
 - 알고리즘 : 주성분 분석(PCA)
 - 문제유형 : 비지도 학습
 - 평가지표 : 기존 분산 반영 비율
 - 사용모델 : KMeans

## 라이브러리 및 데이터 불러오기

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

file_url = 'http://raw.githubusercontent.com/musthave-ML10/data_source/main/customer_pca.csv'
customer = pd.read_csv(file_url)
```

```python
customer.head()
```

 - 변수가 2개인 경우 x, y축에 지정하여 쉽게 그래프를 그릴수 있다
 - PCA를 사용하여 이 많은 변수들을 2개로 축소한 뒤 산점도 그래프로 클러스터 확인해 보자

```python
# PCA 훈련시 종속변수(label)은 제외 시켜야 하므로
customer_X = customer.drop('label', axis=1) # 독립변수 지정
customer_y = customer['label'] # 종속변수 지정
```

### 그래프 표현을 위한 차원 축소

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=2) # 2차원 평면에 그래프를 그리는 것이 목적
```

```python
customer_pca = pca.fit_transform(customer_X) # 학습
```

```python
customer_pca
```

```python
customer_pca = pd.DataFrame(customer_pca, columns=['PC1','PC2'])
```

```python
coustomer_pca = customer_pca.join(customer_y)
```

```python
coustomer_pca.head()
```

```python
sns.scatterplot(x='PC1', y='PC2', data=coustomer_pca, hue='label')
```

 - PCA를 통해 얻어낸 변수 PC1과 PC2는 기존의 모든 변수를 복합적으로 반영하여 만들어짐
 - 명료하게 이 새로운 변수들을 해석하기 쉽지 않음

 - 학습에 사용한 모델 pca 뒤에 components_를 붙여 각 주성분과 기존 변수와의 상관관계 확인 가능

```python
pca.components_
```

```python
df_comp = pd.DataFrame(pca.components_, columns=customer_X.columns)
df_comp
```

 - 인덱스 0과 1은 각각 주성분 PC1과 PC2를 의미

```python
# sns.heatmap(pca.components_) # 가능하지만 변수명이 없어 해석에 어려움
sns.heatmap(df_comp) # 데이터프레임을 활용하므로 변수명까지 포함시킬 수 있음
```

# 데이터의 차원을 축소해 학습 시간을 줄이고 성능을 향상시키자
 - 알고리즘 : 주성분 분석(PCA)
 - 문제유형 : 비지도 학습
 - 평가지표 : AUC
 - 사용모델 : PCA

### 다차원 데이터 불러오기


```python
file_url = 'https://media.githubusercontent.com/media/musthave-ML10/data_source/main/anonymous.csv'
anonymous = pd.read_csv(file_url)
```

```python
anonymous.shape # 독립변수 4296개, 20000개의 데이터
```

```python
anonymous.columns[0] # 종속변수명
```

 - 결측치 확인

```python
# anonymous.isna() # 결측치 → True
# anonymous.isna().sum() # 변수별 (True의)합계 → Series
anonymous.isna().sum().sum() # Series 값의 합계(즉, DataFrame의 결측치 전체합계)
```

### PCA에 따른 모델링 성능/결과 비교하기


```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(anonymous.drop('class', axis=1), anonymous['class'],
                                                    test_size=0.2, random_state=100) # stratify=anonymous['class'] : class의 비율을 train, test에 동일하게 나눠줌
                                                    
```

 - PCA에서도 변수 간의 스케일을 일정하게 맞춰주는 것이 중요 → StandardScaler()

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaler.fit(X_train) # 훈련데이터로 스케일러 훈련

X_train_scaled = scaler.transform(X_train) # 훈련데이터 스케일링
X_test_scaled = scaler.transform(X_test)   # 테스트데이터 스케일링
```

 - PCA를 통한 학습시간 단축이 목표
 - 따라서, 학습시간이 오래걸리는 랜덤 포레스트 모델사용

```python
from sklearn.ensemble import RandomForestClassifier
model_1 = RandomForestClassifier(random_state=100)
```

```python
# 모델 학습에 걸리는 시간 확인(PCA 전 : 변수 4000개)
import time
start_time = time.time()
model_1.fit(X_train_scaled, y_train)
end_time = time.time()
print(end_time-start_time)
```

 - 133초 소요

```python
# 학습된 모델의 정확도 확인
from sklearn.metrics import accuracy_score, roc_auc_score

pred_1 = model_1.predict(X_test_scaled)
accuracy_score(y_test, pred_1) # 정확도
```

 - 96% 정확도

```python
# 학습된 모델의 AUC 확인 : AUC는 소수점 형태로 된 양성 예측값 사용 → predict_proba()
proba_1 = model_1.predict_proba(X_test_scaled)
roc_auc_score(y_test, proba_1[:, 1])
```

 - 0.99 AUC

 - PCA를 통해 4000개의 변수를 시각화를 위한 2개의 변수로 차원축소 : 큰 데이터 손실이 예상됨

```python
pca = PCA(n_components=2)
pca.fit(X_train_scaled)
```

 - pca에 explained_variance_ratio_를 붙여 주성분이 기존 변수의 분산을 얼마나 대변해 주고 있는지 확인
 - 1이면 기존 변수들의 분산을 100% 대변

```python
pca.explained_variance_ratio_
```

 - 주성분 각각 0.0499, 0.0333
 - 둘을 합쳐도 0.08로 기존데이터의 정보를 1%도 반영하지 못한다 : 정보손실이 크다

```python
# 주성분을 몇개로 하는것이 정보손실이 덜 할지 확인
var_ratio = []

for i in range(100, 550, 50): # 100부터 500까지 50개 단위 순회
    pca = PCA(n_components=i)
    pca.fit(X_train_scaled) # 굳이 transform 해줄 필요는 없음
    ratio = pca.explained_variance_ratio_.sum() # array.sum()
    var_ratio.append(ratio)
```

```python
sns.lineplot(x=range(100, 550, 50), y=var_ratio)
```

 - 주성분 수를 100-500까지 확인 결과 : 데이터 반영 비율 62-82%
 - K-평균의 엘보우 기법처럼 드라마틱하게 꺽이는 부분이 없어 적절한 값 찾기 곤란
 - 데이터 손실을 어디까지 감수할 것인지 설정
 - 80%정도의 데이터 반영율을 기준으로 잡아서 주성분이 400개인 데이터를 만들자

```python
pca = PCA(n_components=400, random_state=100)
pca.fit(X_train_scaled) # 훈련데이터를 기준으로 PCA 훈련

X_train_scaled_pca = pca.transform(X_train_scaled) # 훈련데이터 PCA 적용
X_test_scaled_pca = pca.transform(X_test_scaled)   # 시험데이터 PCA 적용
```

```python
# 모델 학습에 걸리는 시간 확인(PCA 후 : 변수400개)
model_2 = RandomForestClassifier(random_state=100)
import time
start_time = time.time()
model_2.fit(X_train_scaled_pca, y_train)
end_time = time.time()
print(end_time-start_time)
```

 - 69초 : 기존 133초 대비 속도 향상

```python
# PCA를 적용한 데이터로 훈련한 모델의 정확도 확인
pred_2 = model_2.predict(X_test_scaled_pca)
accuracy_score(y_test, pred_2)
```

 - 정확도가 거의 유지됨

```python
# PCA를 적용한 데이터로 훈련한 모델의 AUC 확인
proba_2 = model_2.predict_proba(X_test_scaled_pca)
roc_auc_score(y_test, proba_2[:,1])
```

 - PCA 전/후 AUC 점수 차이 없음
